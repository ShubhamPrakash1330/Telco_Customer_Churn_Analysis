### Customer Churn Analysis

```python

import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.info()  # to understand data and its type across columns
```

```python
# to change the value of TotalCharges form "blank" to "0"
df['TotalCharges'] = df['TotalCharges'].replace(" " ,"0")
df['TotalCharges'] = df['TotalCharges'].astype("float")
```

```python
df.info()
```

```python
#  to identify if there is any null value in the dataset
df.isnull().sum()
```

```python
# to understand various dynamics of the dataset
df.describe()
```

```python
df.head()
```

```python
def conv(value):
    if value == 1:
      return "yes"
    else:
      return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(30)
```

```python
# bar chart
```

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
# number of churns
gb= df.groupby('Churn').agg({'Churn':'count'})
gb
```

```python
# Pie chart to represent the percentage of churn
```

```python
plt.figure(figsize =(3,4))
plt.title("Percentage of Churned Customers", fontsize = 10)
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct= "%1.2f%%")
plt.show()
```

```python
# churn on the basis of Gender
ab = sns.countplot(x = 'gender' , data = df , hue = 'Churn')
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn by Gender')
plt.figure(figsize =(3,4))
plt.show()
```

```python
# Churns on the basis of Senior CItizen
ab = sns.countplot(x = 'SeniorCitizen' , data = df, hue = 'Churn' )
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn by Senior Citizen')
plt.figure(figsize =(3,4))
plt.show()
```

```python
# as per partnership status
ab = sns.countplot(x = 'Partner' , data = df, hue = 'Churn' )
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn by Partnership Status')
plt.figure(figsize =(3,4))
plt.show()
```

```python
# as per Dependence
ab = sns.countplot(x = 'Dependents' , data = df, hue = 'Churn' )
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn by Dependents')
plt.figure(figsize =(3,4))
plt.show()
```

```python
# Histogram which is repesenting Churns on the basis of Trenure
plt.figure(figsize=(9,4))
sns.histplot( x= 'tenure' , data = df , bins = 72, hue = 'Churn')
plt.show()
```

### People who  have used  our services for a long time have stayed and people who used our 1 or 2 months are Churned

```python
# Churns on the basis of Contract duration
ab = sns.countplot(x = 'Contract' , data = df , hue = 'Churn')
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn by Contract')
plt.figure(figsize =(3,4))
plt.show()
```

```python
### form the above chart we can conclude how are using servies in month-to-month basis have Churned conparitively then who have used it for 2 years of contracrt
```

```python
df.columns.values
```

```python
# to create multiple bar chart across various columns to identify there relationship with churns
columns_to_plot = ['PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies']


fig, axes = plt.subplots(3, 3, figsize=(12, ))
fig.suptitle('Countplots for Various Features')


for col, ax in zip(columns_to_plot, axes.flatten()):
    sns.countplot(x=col, data=df, ax=ax, hue ='Churn')
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])
    ax.set_title(f'{col} Distribution')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Relationship of Payment Method with churn
plt.figure(figsize =(8,4))
ab = sns.countplot(x = 'PaymentMethod' , data = df , hue = 'Churn')
ab.bar_label(ab.containers[0])
ab.bar_label(ab.containers[1])
plt.title('Churn Customer by Payment Method')
plt.xticks(rotation =30)
plt.show()
```
